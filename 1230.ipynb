```
def enqueue(a, o):
    a.insert(0, o)

def dequeue(a):
    return a.pop()
# A=1 B=2 C=3 D=4 E=5 
g = { #  graph: 
    '1': {'n':['2'], 'v':0}, #  n: neighbor (), v: visited ()
    '2': {'n':['3','4','5'], 'v':0},
    '3': {'n':[], 'v':0},
    '4': {'n':['5'], 'v':0},
    '5': {'n':[], 'v':0},
    
}

def init(g): #   visited  0
    for i in g:
        g[i]['v'] = 0

def dfs(g, node): #  
    if g[node]['v']!=0:           #  
        return
    print(node, '=> ', end = '')  #  
    g[node]['v'] = 1              #    
    neighbors = g[node]['n']      # 
    for n in neighbors:           #  
        dfs(g, n)                 #    

queue=['1'] #  BFS  queue,  1

def bfs(g, q): #  
    if len(q)==0:                 #   queue 
        return
    node = dequeue(q)             #   queue 
    if g[node]['v'] == 0:         #  
        g[node]['v'] = 1          #    
    else:                         #   ()
        return                    #    
    print(node, '=> ', end = '')  #  
    neighbors = g[node]['n']      #  
    for n in neighbors: #  
        if not g[n]['v']:         #  
            enqueue(q, n)         #     queue 
    bfs(g, q)

print('dfs:', end = '')
init(g)
dfs(g, '1') # 
print('')

print('bfs:', end = '')
init(g)
bfs(g, queue) # 
print('')
```
